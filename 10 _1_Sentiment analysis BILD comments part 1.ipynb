The purpose of the following code was to generate sentiment values for each comment based on the tables from notebook 5 (Cleaning the YouTube comments). Due to the large number of comments, the dataset was split into multiple parts, some of which were processed in notebook 10_2 (Sentiment analysis BILD comments part 2). The resulting tables were then merged in notebook 12 (Concatenating BILD comments).
Parts of this code were generated with the help of ChatGPT and altered for the specific needs of this study.

```python
import pandas as pd
from pathlib import Path

BILD_path = Path(r"D:\Users\mague\Desktop\Uni\Bachelor Arbeit\Code\Datasets\Scraped Comments\BILD_comments_full.csv")

BILD_comments = pd.read_csv(BILD_path, encoding="utf-8")

print(BILD_comments.head())
```

```python
BILD_comments = BILD_comments.drop_duplicates().reset_index(drop=True)
print(len(BILD_comments))
```

```python
import numpy as np
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

model_name = "cardiffnlp/twitter-xlm-roberta-base-sentiment"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model=model,
    tokenizer=tokenizer,
    truncation=True,
    max_length=256
)

def preprocess(text):
    return str(text)


def analyze_sentiment(text):
    r = sentiment_analyzer(preprocess(text))[0]
    return r["label"], r["score"]
```

```python
from tqdm import tqdm

tqdm.pandas()

BILD_0_100000 = BILD_comments[:100000]

BILD_0_100000[["sentiment_label", "sentiment_score"]] = (
    BILD_0_100000["text"]
    .progress_apply(lambda x: pd.Series(analyze_sentiment(x)))
)

print(BILD_0_100000[["text", "sentiment_label", "sentiment_score"]].head())

BILD_0_100000.to_csv(r"D:\Users\mague\Desktop\Uni\Bachelor Arbeit\Code\Datasets\Scraped Comments\Comments with sentiment\BILD_0_100000.csv",
          index=False, encoding="utf-8")
```
