The purpose of the following code was to scrape the YouTube URLs of all relevant videos, by using self selected playlist URLs as input. Parts of this code were generated with the help of ChatGPT and altered for the specific needs of this study. The main tool used to scrape the links was YouTube Data API v3.

```python
import re
from googleapiclient.discovery import build

def fetch_playlist_urls(api_key, playlist_url):

    # Extract the playlist_id from the URL
    match = re.search(r"list=([a-zA-Z0-9_-]+)", playlist_url)
    if not match:
        raise ValueError("Invalid YouTube playlist URL. Please provide a valid URL with 'list='.")
    playlist_id = match.group(1)

    # Build the YouTube API client
    youtube = build("youtube", "v3", developerKey=api_key)

    # Initialize list to store video URLs
    video_urls = []
    next_page_token = None

    # Fetch all videos in the playlist
    while True:
        request = youtube.playlistItems().list(
            part="snippet",
            playlistId=playlist_id,
            maxResults=50,
            pageToken=next_page_token
        )
        response = request.execute()

        for item in response["items"]:
            video_url = f'https://www.youtube.com/watch?v={item["snippet"]["resourceId"]["videoId"]}'
            video_urls.append(video_url)

        next_page_token = response.get("nextPageToken")
        if not next_page_token:
            break

    return video_urls
```
