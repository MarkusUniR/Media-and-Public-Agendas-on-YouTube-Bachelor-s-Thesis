The purpose of the following code was to generate sentiment values for each comment based on the tables from notebook 5 (Cleaning the YouTube comments) and the missing comments from notebook 3_2 (Scraping missing comments). Due to the large number of comments, the dataset was split into multiple parts, some of which were processed in notebook 10_1 (Sentiment analysis BILD comments part 1). The resulting tables were then merged in notebook 12 (Concatenating BILD comments).
Parts of this code were generated with the help of ChatGPT and altered for the specific needs of this study.

```python
import pandas as pd
#sentiment classification for the YouTube videos' title and description

BILD_comments = pd.read_csv("BILD_comments_full.csv", encoding="utf-8")

BILD_comments = BILD_comments.drop_duplicates().reset_index(drop=True)

print(BILD_comments.shape)
```

```python
import pandas as pd
import numpy as np
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

model_name = "cardiffnlp/twitter-xlm-roberta-base-sentiment"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model=model,
    tokenizer=tokenizer,
    truncation=True,
    max_length=256
)

def preprocess(text):
    return str(text)


def analyze_sentiment(text):
    r = sentiment_analyzer(preprocess(text))[0]
    return r["label"], r["score"]

```

```python
from tqdm import tqdm


tqdm.pandas()

BILD_1000000_2000000 = BILD_comments[1000000:2000000]

BILD_1000000_2000000[["sentiment_label", "sentiment_score"]] = (
    BILD_1000000_2000000["text"]
    .progress_apply(lambda x: pd.Series(analyze_sentiment(x)))
)

print(BILD_1000000_2000000[["text", "sentiment_label", "sentiment_score"]].head())
```

```python
BILD_1000000_2000000.to_csv("BILD_1000000_2000000.csv",
          index=False, encoding="utf-8")
```

```python
from tqdm import tqdm

tqdm.pandas()

BILD_2000000_3000000 = BILD_comments[2000000:3000000]

BILD_2000000_3000000[["sentiment_label", "sentiment_score"]] = (
    BILD_2000000_3000000["text"]
    .progress_apply(lambda x: pd.Series(analyze_sentiment(x)))
)

print(BILD_2000000_3000000[["text", "sentiment_label", "sentiment_score"]].head())
```

Some of the comments weren't properly scraped which is why they are analysed seperately here. Along with the changes that had to be done to make the users anonymous.

```python
import pandas as pd
BILD_missing_comments = pd.read_csv("missing_BILD_comments.csv", encoding="utf-8")

BILD_missing_comments = BILD_missing_comments.drop_duplicates().reset_index(drop=True)

# drop unneccessary columns
BILD_missing_comments = BILD_missing_comments.drop(columns=["author", "updated_at"], errors="ignore")

# make commentors anonymous
BILD_missing_comments["text"] = (
    BILD_missing_comments["text"]
    .astype(str)
    .str.replace(r"@\S+", "", regex=True)
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)
```

```python
import emoji
import pandas as pd

# converting the emojis in the comments into text for later sentiment analysis
def demojize_german(text):
    if not isinstance(text, str):
        return text

    return emoji.demojize(text, language="de")

BILD_missing_comments["text"] = BILD_missing_comments["text"].apply(demojize_german)
```

```python
from tqdm import tqdm

tqdm.pandas()

BILD_missing_comments[["sentiment_label", "sentiment_score"]] = (
    BILD_missing_comments["text"]
    .progress_apply(lambda x: pd.Series(analyze_sentiment(x)))
)
print(BILD_missing_comments[["text", "sentiment_label", "sentiment_score"]].head())

from google.colab import files

BILD_missing_comments.to_csv("BILD_missing_comments_with_sentiment.csv",
          index=False, encoding="utf-8")
files.download("BILD_missing_comments_with_sentiment.csv")
```
