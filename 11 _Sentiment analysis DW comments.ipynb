The purpose of the following code was to generate sentiment values for each DW comment based on the tables from notebook 5 (Cleaning the YouTube comments).
Parts of this code were generated with the help of ChatGPT and altered for the specific needs of this study.

```python
import pandas as pd
DW_comments = pd.read_csv("DW_comments_full.csv", encoding="utf-8")
DW_comments = DW_comments.drop_duplicates().reset_index(drop=True)
```

```python
DW_comments
```

```python
import numpy as np
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

model_name = "cardiffnlp/twitter-xlm-roberta-base-sentiment"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model=model,
    tokenizer=tokenizer,
    truncation=True,
    max_length=256
)

def preprocess(text):
    return str(text)


def analyze_sentiment(text):
    r = sentiment_analyzer(preprocess(text))[0]
    return r["label"], r["score"]
```

```python
from tqdm import tqdm

tqdm.pandas()

DW_comments[["sentiment_label", "sentiment_score"]] = (
    DW_comments["text"]
    .progress_apply(lambda x: pd.Series(analyze_sentiment(x)))
)
DW_comments.to_csv("DW_comments_final.csv", index=False)
```
